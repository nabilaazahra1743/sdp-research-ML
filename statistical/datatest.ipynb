```python
from scipy import stats
import numpy as np
import pandas as pd

data2 = pd.read_csv('D:\AAAAKK SKRIPSWEETT\Bismillah Code Skripsi V2\Dataset\dataset_sdp.csv', delimiter=";")

z_scores = np.abs(stats.zscore(data2['nilai']))
data2['z_score'] = z_scores

outliers_z = data2[data2['z_score'] > 3]  # Umumnya z > 3 dianggap outlier
print("Outliers via Z-Score:\n", outliers_z)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Anggap kolom 'bugs' sudah bernilai 0 (non-bug) dan 1 (bug)
sns.countplot(x='bugs', data=data2)
plt.title("Data Distribution of SDP Dataset")
plt.xlabel("0=Non-Bug, 1=Bug")
plt.ylabel("count")
plt.show()
```
